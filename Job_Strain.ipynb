```python
import pandas as pd
df = pd.read_csv('Job_Strain_2015.csv')
print(df)
```

```python
df_new = df.pivot_table(values = 'Value', index = 'Country', columns = 'Components')
df_new.reset_index(inplace=True)
df_new.columns = df_new.columns.str.replace(' ', '_')
half_df_new = df_new.head(18)
```

```python
import matplotlib.pyplot as plt
half_df_new.plot.bar(x='Country', y=['Inflexibility_of_working_hours', 'Job_Strain', 'Long_working_hours','Opportunity_for_career_advancement', 'Physical_health_risk_factors', 'Training_and_learning', 'Work_autonomy_and_learning_opportunities'], cmap = 'viridis', rot=0)
plt.rcParams["figure.figsize"] = (15, 10)
plt.xticks(rotation = 'vertical')
plt.title('Job Strain Measures of 2005, 2010, and 2015', fontsize=16)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Job Strain Values', fontsize=16)
plt.savefig('Job_Strain_measures_AUS-JPN.jpeg', bbox_inches = 'tight')
plt.show()
```

```python
import seaborn as sb
#Time against Value
ax = sb.barplot(x="Time", y="Value", hue="POP", data=df, palette = "viridis")
plt.title('Job Strain of Different Age Group at Different Year', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.savefig('Job_Strain_Age_Year.jpeg', bbox_inches = 'tight')
plt.show()
```

```python
ax = sb.barplot(x="Time", y="Value", hue="Sex", data=df, palette = "viridis")
plt.title('Job Strain of Different Gender at Different Year', fontsize=16)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.savefig('Job_Strain_Gender_Year.jpeg', bbox_inches = 'tight')
plt.show()
```

```python
ax = sb.barplot(x="Country", y="Value", hue="Sex", data=df)
plt.title('Different age group at different type value Job Strain')
plt.xticks(rotation = 'vertical')
plt.show()
```
